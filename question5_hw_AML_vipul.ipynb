```python
#reading file into data frame
import pandas as pd
header_data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.names', skiprows = 33, error_bad_lines=False, header=None)
header_data
df= pd.DataFrame(header_data)
df
```

```python
# separating data to fetch the coloums and further appended target to data set 
coloumn = [x for x in df[0]]

dff=[]

for i in range(0,len(coloumn)):
    m = coloumn[i].split(':')[0]
    dff.append(m)
    z=0
dff.append('target')
# dff        
dff                        
```

```python
# added coloums to the given data set and futher changed the value greater than zero to 1 
#http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data
header_data2 = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data',error_bad_lines=False, header=None)
header_data2.columns=dff
for z in dff:
    header_data2.loc[header_data2[z] > 0, z] = 1
    
header_data2    
```

```python

#y is to be predicted ie target variable 
#x is explanatory variable 
from sklearn.model_selection import train_test_split
y = header_data2.target
x = header_data2.drop('target',axis=1)
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)

```

```python
# printed calulation of confusion matrix usingg alpha = 0 
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix

bnb = BernoulliNB(alpha = 0)
model = bnb.fit(x_train,y_train)

y_pred = model.predict(x_test)

C=confusion_matrix(y_test, y_pred)


```
